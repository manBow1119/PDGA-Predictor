## Combining All Tourneys into One Dataframe

```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# import CSVs to dataframes for concatenation
LVC = pd.read_csv('Tournament_CSVs/LVC_22.csv')
DDO = pd.read_csv('Tournament_CSVs/DDO_final.csv')
DGLO = pd.read_csv('Tournament_CSVs/DGLO_22.csv')
EURO = pd.read_csv('Tournament_CSVs/EURO_22.csv')
EURO = pd.read_csv('Tournament_CSVs/EURO_22.csv')
EURO.head()
```

```python
DGLO['Round_4'] = DGLO[['Round_1', 'Round_2', 'Round_3']].mean(axis = 1).astype(int)
```

```python
DGLO = DGLO[['Place','Tournament', 'Name', 'Round_1', 'Round_2', 'Round_3', 'Round_4', 'Payout']]
DGLO.head()
```

```python
EURO['Round_4'] = EURO[['Round_1', 'Round_2', 'Round_3']].mean(axis = 1).astype(int)
EURO = EURO[['Place','Tournament', 'Name', 'Round_1', 'Round_2', 'Round_3', 'Round_4', 'Payout']]
EURO.head()
```

```python
# Concatenate all dfs into one giant df
frames = [LVC, DDO, DGLO, EURO]
all_df = pd.concat(frames)
all_df.tail()
```

```python
all_df.set_index('Place', inplace = True)
all_df.head()
```
